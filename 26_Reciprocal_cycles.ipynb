A unit fraction contains 1 in the numerator. The decimal representation of the unit fractions with denominators 2 to 10 are given:

1/2	= 	0.5<br />
1/3	= 	0.(3)<br />
1/4	= 	0.25<br />
1/5	= 	0.2<br />
1/6	= 	0.1(6)<br />
1/7	= 	0.(142857)<br />
1/8	= 	0.125<br />
1/9	= 	0.(1)<br />
1/10	= 	0.1<br />
Where 0.1(6) means 0.166666..., and has a 1-digit recurring cycle. It can be seen that 1/7 has a 6-digit recurring cycle.

Find the value of d < 1000 for which 1/d contains the longest recurring cycle in its decimal fraction part.

```python
import decimal
mynum = decimal.Decimal(3)
mynum = 1/mynum
```

```python
mynum
```

```python
mydecs = [(0,0)] # initialized with 0 so index matches denominator
decimal.getcontext().prec = 3100 # could be up to 3000 non-repeating characters before repeat for 3-digit divisors
for i in range(1,1001):
    mydecs.append((i,1 / decimal.Decimal(i)))
```

```python
len(str(mydecs[3][1]))
```

```python
myrepdecs = []
for dec in mydecs:
    if len(str(dec[1])) >= 3000: # strip out non-repeating chars
        myrepdecs.append(dec)
```

```python
len(myrepdecs)
```

```python
repdecslens = []
for repdec in myrepdecs:
    chunk = ''
    repstr = str(repdec[1])[50:]
    for i in range(0,len(repstr)): # get rid of part that might not be repeating chunk (doesn't matter where we start)
        chunk += repstr[i]
        if chunk + chunk == repstr[:len(chunk)*2] and len(chunk) >= 3: # d < 1000 means repeating chunks < 3 digs aren't full repeat
            repdecslens.append((repdec[0],repdec[1],chunk))
            break
```

```python
biggest = 0
d = 0
for item in repdecslens:
    if len(item[2]) > biggest:
        biggest = len(item[2])
        d = item[0]

print(f'1/{d} has repeating pattern of {biggest} length')
```
